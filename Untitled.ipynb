```python
train.damage_grade = train.damage_grade.str.slice(-1)
train.damage_grade=train.damage_grade.astype(int)
```

```python

train.groupby('district_id',as_index=False).sum().sort_values(by='damage_grade')
```

```python
train['summ'] = train.groupby('district_id',as_index=False).sum()[['damage_grade']]
#train.district_id = train.groupby('district_id')['damage_grade'].transform('sum')
train.summ.describe()
```

```python
#dividing disctricts into  different bins at points(2395,17160,34512,126288,249063)
train.district_id = pd.cut(train['summ'],bins=[2395,17160,34512,126288,249063],labels=['dis_catagory_1','dis_catagory_2','dis_catagory_3','dis_catagory_4'])
```

```python
train=train.drop('vdcmun_id',axis=1) 
train=train.drop('building_id',axis=1)
train.head(20)
```

```python
train.fillna(method='ffill',inplace=True)
```

```python

y = train.damage_grade
x= train.drop('damage_grade',axis=1)
x= train.drop('summ',axis=1)
```

```python
x = pd.get_dummies(x)
```

```python
x.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
train_x, val_x, train_y, val_y = train_test_split(x,y,random_state = 0)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(penalty='l2', C=1)
```

```python
model.fit(train_x,train_y)
```

```python
model.predict(val_x)
```

```python
print("Logistic accruracy %2.2f "%accuracy_score(val_y,model.predict(val_x)))
```

```python
from sklearn.tree import DecisionTreeClassifier

# Define model
train_model = DecisionTreeClassifier()
train_model.fit(train_x,train_y)
```

```python
train_model.score(val_x, val_y)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(train_x,train_y)
```

```python
model.score(val_x, val_y)
```
